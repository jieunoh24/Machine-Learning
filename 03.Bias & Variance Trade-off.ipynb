## 3. Bias & Variance Trade-off
---
- bias: 우리가 적합한 머신러닝 모델이 실제값과 얼마나 닮았는가를 나타내는 지표(goodness-of-fit)
- variance: 모델이 우리가 훈련시킨 데이터셋 이외의 데이터셋에도 얼마나 잘 들어맞는지에 대한 지표   
=> 일반적으로 모델이 복잡해질수록 높아지는 경향이 있음

- 1번 그림의 경우 모델이 매우 심플하기 때문에 variance는 작지만, 실제값과 모델간의 차이가 상당하므로 높은 bias를 가지고 있음 -> underfitting 되었다고 할 수 있음.
- 3번 그림의 경우 모델이 실제값과 매우 잘 적합하여 낮은 bias를 가지지만, 모델이 상당히 복잡하여 높은 variance를 가짐 -> overfitting 되었다고 할 수 있음.

- bias와 variance 사이에는 트레이드 오프 관계가 존재하기 때문에, underfitting과 overfitting을 최소화할 수 있는 최적의 모델을 찾는 것이 목표임.
- Bias와 Variance를 수식으로 나타내기

- f(x)가 복잡할수록 더 많은 데이터를 포착할 수 있기 때문에 Bias가 작아지지만, 각 점을 포착하기 위해 더 많이 움직여야 하므로 Variance가 커짐.
