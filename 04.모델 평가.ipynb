## 4. 모델 평가
---
1. 성능 평가 방법

위의 그림처럼 데이터셋을 Train / Validation / Test set으로 나눌 수 있음.

1) Train set             
train 데이터셋은 모델을 훈련시킬 때 사용됨. 이때 모델을 너무 복잡하게 만든다면 overfitting이 발생할 수 있으므로 유의해야함.                
2) Validation set            
validation 데이터셋은 우리가 train set에서 적합한 모델의 parameter를 수정하거나, overfitting 여부를 확인하는데 사용됨                    
3) Test set
test 데이터셋은 우리가 앞서 적합한 모델의 성능을 평가하는데 사용됨.     

---
- 데이터 수가 적다면 데이터 분할이 좋은 방법이 아닐 수 있음             
=> 이에 대한 대안으로 여러가지 방법의 모델 평가 방법이 존재함             
1) k-fold cross validation

데이터셋을 k개로 분할한 후 k개의 데이터 조각 중에서 k-1개를 train set으로 지정하고 남은 한 개의 데이터 조각을 test set으로 지정하여 학습하는 방법임. k번의 평가 후 평가 지표의 평균을 결과로 정함.        
2) Bootstrapping                 

원래의 데이터셋에서 m개의 데이터를 랜덤으로 추출하여 데이터 샘플을 B개 생성함. B개의 데이터 샘플로 각각 B개의 모델을 만들고 test set은 원래 데이터셋으로 하여 평가를 진행함. 그리고 B개의 평가 지표의 평균을 결과로 저장함.
---
2. 성능 평가 지표
- 성능 평가 지표는 회귀 모형과 분류 모형에 따라 달라짐        
(1) 회귀 모형의 성능을 평가하는 지표                 
- SSE(Sum of squared error):              
실제값과 예측값의 차이를 제곱하여 전부 더한 값 -> 실제로 회귀 모형을 평가할 때 자주 쓰이는 지표

- 결정 계수(R-squared):         
선형 회귀 분석의 성능 검증 지표로 많이 이용됨. 적합한 회귀 모형이 실제값을 얼마나 잘 적합하는지에 대한 비율

- MAE(Mean absolute error):                
오차의 절대값을 평균낸 값

- MAPE(Mean absolute percentage error):      
MAE를 계산할 때 실제값에 대한 상대적인 비율을 고려하여 계산된 값



(2) 분류 모형의 성능을 평가하는 지표
- Confusion matrix(혼돈 행렬): 분류 모형의 성능은 혼돈 행렬을 통해 평가될 수 있음


- Accuracy:     
모든 예측값들 중에서 실제 값을 맞춘 개수의 비율

- Precision:             
긍정으로 예측한 모든 값 중에서 실제값도 긍정인 값의 비율

- Recall:        
긍정을 맞춘 개수를 정답을 맞춘 개수로 나눈 값. 정답 중에서 긍정을 맞춘 개수의 비율

- Specificity:            
부정을 맞춘 개수를 실제 부정의 개수로 나눈 값. 실제 부정 중에서 부정으로 예측된 개수의 비율

- F1 Score:       
Precision과 Recall의 조화평균. F1 스코어는 주로 데이터의 label이 불균형하게 분포할 때의 성능 평가 지표로 사용됨. 예를들어, positive가 1000개, negative가 50개밖에 없다면, 1050개 모두 positive로 예측해도 accuracy는 높은 수치로 보이게됨. 이런 문제에 대비해 f1 스코어를 사용함. 

- ROC 곡선:            
가로축을 False Positive Rate(Specificty)로 하고, 세로축을 True Positive Rate(Sensitivity)로 하여 시각화한 그래프. 그래프가 위로 갈수록 좋은 모델이라고 해석할 수 있음. 적어도 y=x 그래프보다 위에 있어서 어느정도 유의미한 모델로 볼 수 있음.
